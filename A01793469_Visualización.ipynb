#***Descripción de las medidas de Tendencia Central y Dispersión***
##*count(Conteo): Es la cantidad de datos diferentes a NA que tiene cada columna.*
##*mean(Promedio/Media): Es la suma de todos los datos distintos a NA de cada columna, dividido entre la cantidad de datos distintos a NA.*
##*std(Desviación Estandar): Es una métrica que da una idea de qué tan dispersos o alejados se encuentran los datos del conjunto respecto del promedio de cada columna.*
##*min(Mínimo): Es el valor mínimo de cada columna.*
##*25%(Primer cuartil): Ordenando el conjunto de menor a mayor, este es el valor que se encuentra en el lugar 0.25*count de cada columna, si el resultado es decimal, se hace una interpolación entre los datos con las posiciones de los enteros inmediatos anterior y siguiente al número obtenido.*
##*50%(Mediana): Ordenando el conjunto de menor a mayor, este es el valor que se encuentra en el lugar count/2 de cada columna, si el resultado es decimal, se obtiene el promedio de los números con las posiciones inmediatamente anterior y siguiente del número obtenido.*
##*75%(Tercer cuartil): Ordenando el conjunto de menor a mayor, este es el valor que se encuentra en el lugar 0.75*count de cada columna, si el resultado es decimal, se hace una interpolación entre los datos con las posiciones de los enteros inmediatos anterior y siguiente al número obtenido.*
##*max(Máximo): Es el valor máximo de cada columna.*

##**10. Interpreta y explica cada uno de los gráficos indicando cuál es la información más relevante que podría ayudar en el proceso de toma de decisiones.**

#### *Primer gráfico: El primer gráfico es el de dispersión (scatter), con el cuál podemos meter dos variales diferentes, que cada una represente un eje de un plano, y poder observar cómo se relacionan ambas. Lo más importante de identificar en este tipo de gráficos es la correlación, es decir, si los puntos parecen formar una recta o estar muy cerca de una, la correlación es muy cercana a 1, o si los puntos se encuentran muy dispersos, sin acercarse a parecer formar una recta, la correlación tiende a 0. En este caso observamos la dispersión entre la "Cantidad de crédito otorgada (dólares)" a cada usuario y el monto en el "Estado de cuenta de Septiembre (dólares)" y como se puede apreciar, los puntos no se acercan a formar una recta, entonces podemos concluir que la correlación entre ambas variables es muy cercana a 0. La correlación es útil para poder establecer relaciones entre variables, y, en caso de tener una correlación muy cercana a 1, se puede reducir una de ambas variables, ya que la información que dan es relativamente "repetitiva".*
#### *Segundo gráfico: Este consiste en un gráfico de caja y bigotes (boxplot), en esta se puede observar la distribución de los valores en cuartiles, es decir, si los ordenamos de menor a mayor, cuáles son los valores que se encuentran en el 25%, 50% y 75% de la cantidad de datos. En mi caso, realicé el gráfico con la variable "Edad (años)", podemos apreciar que el 75% de los usuarios tienen entre 20 y 40 años, a pesar de que la edad máxima es de casi 80 años, es decir, el último 25% tiene una edad de entre 40 y 80 años. Esto es importante para saber qué rangos de datos son los que tienen mayor peso dentro de todo el grupo.*
#### *Tercer gráfico: El tercer gráfico es una gráfica de pastel (pie), esta nos ayuda principalmente a identificar la frecuencia de valores dentro de un conjunto de datos, debido a esta naturalesa, su función se aprovecha de la mejor manera para datos de tipo categóricos. En nuestro caso, realizamos este gráfico para la columna "Género" obteniendo los porcentajes de usuarios que son Mujeres (60.4%) y Hombres (39.6%). Esta información es útil para observar qué la proporción que exista entre las distintas categorías de un conjunto, identificando cuáles son los que tienen mayor presencia.*
#### *Cuarto gráfico: El cuarto y último gráfico es una gráfica de barras (bar), esta puede ser utilizada con tipos de datos numéricos y categóricos, en cada uno de ellos tiene un objetivo diferente. Para los categóricos, como fue en este caso, sirve, de manera análoga al gráfico de pastel, para identificar la frecuencia de los datos dentro de un conjunto, sin embargo, en este tipo de gráfico se manejan mayormente los valores de frecuencia (cantidad de veces que ocurre) y no como porcentaje del total. Yo usé este gráfico con la variable "Estado Civil", donde podemos apreciar que el de mayor frecuencia son los usuarios Solteros, seguidos de los Casados. Esto nos ayuda de la misma manera que en el caso de la gráfica de pastel, identificando qué grupos son los más presentes dentro de un conjunto.*
