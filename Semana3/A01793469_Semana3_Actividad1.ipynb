##Fundamentos de bases de datos y para ciencia de datos.
**Las bases de datos son herramientas muy poderosas en cuanto a la ciencia y manejo de datos concierne, las cuales consisten de tablas con columnas y renglones que poseen datos de manera organizada. Su gran utilidad es debido a que son eficientes al momento de almacenar y manejar la información o datos que se utilizan dentro de diferentes proyectos. Dentro de estas, existen 2 tipos: las relacionales, que como su nombre lo indica, permiten establecer relaciones entre las diferentes tablas, permitiendo un mejor entendimiento del modelo; y las no relacionales, que buscan desplegar toda la información de una manera más compacta y abstracta, definiendo cada valor almacenado a una llave única e irrepetible. Personalmente he tenido la oportunidad de trabajar con bases de datos, del tipo SQL, que utilizan las relacionales y me parece de gran utilidad poder establecer relaciones entre las diferentes tablas, logrando así mayor estructura, alcance y organización dentro de la misma base de datos.**
##Fundamentos de almacenes de datos (Data Warehouse) para ciencia de datos
**El "Data Warehouse" es un concepto bastante llamativo para mi, ya que yo no consideraba que existieran sitios o servidores específicamente dedicados únicamente a almacenar y manejar bases de datos. A priori, esta tarea puede parecer sencilla, en el aspecto que uno podría pensar que no hay dificultad en solamente introducir datos a un servidor, lo cual es en parte correcto. El reto entra al momento de introducir los datos de manera organizada, bajo el mismo formato, sin ruido o datos corruptos; saber determinar qué información es relevante y cuál no, establecer la relación entre diferentes datos, definir los procesos para trabajar, refinar, purificar y almacenar los datos, todo esto requiere de una habilidad y criterio bastante desarrollados, ya que las empresas utilizan los resultados del procesamiento de estos datos para tomar decisiones. Existen 2 tipos de procesos para almacenar datos provenientes de diferentes fuentes en un Warehouse. El primero es ETL(Extract-Transform-Load) en el cual se obtienen los datos de una fuente específica, se transforman de manera externa a la base de datos objetivo, se procesan y refinan los datos de cierta forma y posteriormente se suben al Warehouse. El segundo es el ETL(Extract-Load-Transform), en donde podemos apreciar que se invierten 2 pasos, se extrae de la misma manera la información, sin embargo se carga directamente al Warehouse, sin procesamiento, y ya dentro de él, con sus recursos, se procesa la información para dejarla lista para ser utilizada, analizada o estudiada.**
